# Functions
This notebook just contains some common functions for discussion in class.

#### Libraries
Numpy is a Python library for handling numerical arrays and operations.
It is very efficient at dealing with large matrices and sets of data.

Matplotlib is a plotting library for Python that works with numpy.

```python
import numpy as np
import matplotlib.pyplot as pl
```

The built-in `range()` function in Python creates an array (technically, generates) of integer values in progression.
Numpy's `arange()` function is similar, but supports floating point numbers.

```python
x = np.arange(-10., 10., 0.01)
```

#### Linear
Linear relationships in one variable are of the form $y = mx + c$. 

```python
pl.plot(x,  5.0 * x + 7.0, 'b', label=r'$5x + 7$')
pl.plot(x,  10.0 * x + 7.0, 'g', label=r'$10x + 7$')
pl.plot(x,  -5.0 * x + 7.0, 'r', label=r'$-5x + 7$')
pl.legend()
pl.show()
```

```python
pl.plot(x,  5.0 * x + 7.0, 'b', label=r'$5x + 7$')
pl.plot(x,  5.0 * x + 14.0, 'g', label=r'$5x + 14$')
pl.plot(x,  5.0 * x + 21.0, 'r', label=r'$5x + 21$')
pl.legend()
pl.show()
```

#### Powers
In Python, a double asterisk means *to the power of*.
A single asterisk means multiply, as usual.

```python
y = x**2
pl.plot(x, y, label=r'$x^2$')
pl.legend()
pl.show()
```

```python
y = x**3
pl.plot(x, y, label=r'$x^3$')
pl.legend()
pl.show()
```

```python
y = x**4
pl.plot(x, y, label=r'$x^4$')
pl.legend()
pl.show()
```

```python
y = x**5
pl.plot(x, y, label=r'$x^5$')
pl.legend()
pl.show()
```

```python
y = x**6
pl.plot(x, y, label=r'$x^6$')
pl.legend()
pl.show()
```

```python
y = x**7
pl.plot(x, y, label=r'$x^7$')
pl.legend()
pl.show()
```

```python
y = np.exp(x)
pl.plot(x, y, label=r'$e^2$')
pl.legend()
pl.show()
```

```python
y = np.sqrt(x[x >= 0])
pl.plot(x[x >= 0], y, label=r'$\sqrt{x}$')
pl.legend()
pl.show()
```

```python
y = x**2 + 100
pl.plot(x, y, label=r'$x^2 + 100$')
pl.legend()
pl.show()
```

```python
y = x**2 - 100
pl.plot(x, y, label=r'$x^2 - 100$')
pl.legend()
pl.show()
```

```python
y = x**2 - 5*x + 4
pl.plot(x, y, label=r'$x^2 - 5x + 4$')
pl.legend()
pl.show()
```

```python
y = np.tanh(x)
pl.plot(x, y, label=r'$\tanh(x)$')
pl.legend()
pl.show()
```

```python
y = 1 / (1 + np.exp(-x))
pl.plot(x, y, label=r'$S(x)$')
pl.legend()
pl.show()
```
